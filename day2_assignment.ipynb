## Task 1 — Your step count
Here are your daily step counts for one week:
`4000, 6500, 3000, 8000, 7200, 5000, 9000`

1. Store them in an array called `steps`.
2. Print the total, average, highest, and lowest steps.
3. Print the standard deviation (rounded to 2 decimals).

````python
# YOUR CODE HERE
steps = np.array([4000, 6500, 3000, 8000, 7200, 5000, 9000])

print("average steps: ", steps.mean())
print("highest steps: ", steps.max())
print("lowest steps: ", steps.min())

print("Standardized deviation: ", steps.std())
````

## Task 2 — Goal check with masks
Your daily step goal is **6000**.
1. Use a boolean mask to show only the days you hit the goal (≥ 6000).
2. Count how many days you hit the goal.
3. Show the days you missed it.

````python
# YOUR CODE HERE

print(steps[steps >= 6000])
print(sum(steps >= 6000))
print(steps[steps < 6000])
````

## Task 3 — Label your days with np.where
Create a label array from `steps`:
- `"GOAL"` if steps ≥ 6000
- `"MISS"` otherwise

Print the labels.

````python
# YOUR CODE HERE

print(np.where(steps >= 6000 , "GOAL" , "MISS"))
````

## Task 4 — Work with a table (axis)
This 2D array shows 3 weeks × 4 subjects of quiz scores:

```
Week1: 8, 7, 9, 6
Week2: 5, 8, 7, 9
Week3: 9, 9, 8, 7
```

1. Build the 2D array called `quiz`.
2. Print each week's average (across rows).
3. Print each subject's average (down columns).

````python
# YOUR CODE HERE
import numpy as np

quiz = np.array([[8, 7, 9, 6],
                [5, 8, 7, 9],
                [9, 9, 8, 7]])

print("week's row's average: ", quiz.mean(axis = 1))
print("week's colum's average: ", quiz.mean(axis = 0))
````
